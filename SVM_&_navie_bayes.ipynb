1.What is a Support Vector Machine (SVM)


--->A **Support Vector Machine (SVM)** is a supervised machine learning algorithm used mainly for classification, where it finds the best boundary (hyperplane) that separates different classes with the maximum margin. The data points closest to this boundary are called **support vectors**, and they determine how the hyperplane is positioned. SVM can also handle non-linear data using **kernel functions** like RBF or polynomial, which transform the data into higher dimensions so it becomes separable. It is powerful, accurate, and works well especially with high-dimensional or complex datasets.


2.What is the difference between Hard Margin and Soft Margin SVM

--->**Hard Margin SVM** allows *no misclassification* at all. It draws a separating hyperplane such that **all data points lie outside or exactly on the margin**, meaning the data must be *perfectly linearly separable*. It is very strict and sensitive to noise or outliers.
**Soft Margin SVM**, on the other hand, allows **some misclassification** by introducing slack variables. It balances between maximizing the margin and minimizing classification errors, making it more flexible, more realistic, and suitable for real-world noisy datasets.


3. What is the mathematical intuition behind SVM

--->The **mathematical intuition behind SVM** is that it tries to find a decision boundary (hyperplane) that **maximizes the margin** between two classes. The margin is the distance from the hyperplane to the nearest data points (support vectors). Mathematically, SVM tries to **maximize this distance** while still correctly classifying the points. This becomes an optimization problem:

* It minimizes the **weight vector** ( ||w||^2 ) (which maximizes the margin).
* Subject to constraints that each point is correctly classified:
  [
  y_i (w \cdot x_i + b) \ge 1
  ]
* For soft margin, slack variables ( \xi_i ) allow some violations.

In simple terms, SVM chooses the hyperplane that is **as far as possible from both classes**, making the model more robust and reducing overfitting.


4.What is the role of Lagrange Multipliers in SVM

-->In SVM, **Lagrange Multipliers** help convert the constrained optimization problem (maximize margin while correctly classifying points) into a **dual problem** that is easier to solve mathematically and computationally.

### ðŸ”¹ Role of Lagrange Multipliers in SVM

1. **Convert constraints into the objective function**
   SVM has constraints like
   [
   y_i(w \cdot x_i + b) \ge 1
   ]
   Using Lagrange multipliers ( \alpha_i ), these constraints are incorporated into a single function called the **Lagrangian**.

2. **Transform the primal problem into the dual problem**
   The dual form makes training simpler and allows the use of **kernels**.
   You end up optimizing in terms of the multipliers ( \alpha_i ), not ( w ) and ( b ).

3. **Identify Support Vectors**
   Only data points with **Î± > 0** become support vectors.
   These points lie **on or inside the margin** and determine the hyperplane.
   Points with **Î± = 0** do not affect the boundary.

4. **Enable use of Kernel Trick**
   In the dual problem, data appear only in dot products.
   This allows replacing dot products with **kernel functions**, enabling non-linear SVM.



5.What are Support Vectors in SVM

--->**Support Vectors** are the **data points that lie closest to the decision boundary (hyperplane)** in a Support Vector Machine. They are the most critical points because:

### ðŸ”¹ **They determine the position and orientation of the hyperplane**

Removing or changing these points will change the boundary.

### ðŸ”¹ **They are the points with non-zero Lagrange multipliers (Î± > 0)**

These points lie **on the margin** or **inside it** (in soft-margin SVM).

### ðŸ”¹ **They maximize the margin**

The margin is measured directly using these points.




6.What is a Support Vector Classifier (SVC)

--->A **Support Vector Classifier (SVC)** is the practical implementation of SVM used for **classification tasks**. It finds the best separating hyperplane between classes by maximizing the margin while allowing some misclassification (soft margin). SVC can also handle **non-linear data** using kernel functions like RBF, polynomial, or sigmoid. In simple terms, SVC is the classifier version of SVM that learns the optimal boundary to separate different classes.


7.What is a Support Vector Regressor (SVR)

--->A **Support Vector Regressor (SVR)** is the regression version of Support Vector Machines. Instead of trying to classify data into categories, SVR tries to **fit a curve or line** that predicts continuous values while keeping the errors within an acceptable range. It uses an **Îµ-insensitive margin**, meaning errors smaller than Îµ are ignored. SVR aims to find a function that is as flat as possible while allowing some points to lie outside the Îµ margin using slack variables. In short, **SVR predicts continuous values by finding a balanced, margin-based fit that is robust to outliers and noise.**


8.What is the Kernel Trick in SVM

--->The **Kernel Trick** in SVM is a mathematical technique that allows the model to handle **non-linear data** without explicitly transforming it into a higher-dimensional space. Instead of performing the actual transformation, SVM uses a **kernel function** to compute the dot product of data points **as if** they were transformed.

### ðŸ”¹ Why is it useful?

* Many datasets are **not linearly separable** in their original form.
* Mapping them to a higher-dimensional space makes them separable.
* But explicitly transforming data is **computationally expensive**.

### ðŸ”¹ How the kernel trick helps:

It replaces the dot product ( x_i \cdot x_j ) with a kernel function ( K(x_i, x_j) ), such as:

* **RBF (Gaussian) Kernel**
* **Polynomial Kernel**
* **Sigmoid Kernel**
* **Linear Kernel**

10.What is the effect of the C parameter in SVM

--->The **C parameter** in SVM controls the trade-off between **maximizing the margin** and **minimizing classification errors**.

---

## ðŸ”¹ **Effect of C**

### **1. Small C (low penalty)**

* Allows more misclassifications
* Focuses on maximizing the margin
* Creates a **wider margin**
* Leads to a **simpler, smoother** decision boundary
* Lower risk of overfitting

### **2. Large C (high penalty)**

* Strongly penalizes misclassification
* Forces the model to classify every training point correctly
* Creates a **narrower margin**
* Fits more tightly around the data
* Higher risk of **overfitting**




11.What is the role of the Gamma parameter in RBF Kernel SVM

--->In an **RBF Kernel SVM**, the **Gamma (Î³)** parameter controls **how far the influence of a single training point reaches**.

---

## ðŸ”¹ **Effect of Gamma**

### **1. Small Gamma (low Î³)**

* Each data point has a **large radius of influence**
* Decision boundary becomes **smooth and simple**
* Model is **less sensitive** to individual points
* Lower risk of overfitting

### **2. Large Gamma (high Î³)**

* Each data point has a **small radius of influence**
* Decision boundary becomes **very wiggly and complex**
* Model becomes **highly sensitive** to individual points
* Higher risk of **overfitting**



12. What is the NaÃ¯ve Bayes classifier, and why is it called "NaÃ¯ve"

--->A **NaÃ¯ve Bayes classifier** is a simple probabilistic machine learning algorithm based on **Bayesâ€™ Theorem**, used mainly for classification tasks like spam detection, sentiment analysis, and document categorization. It calculates the probability of each class given the input features and chooses the class with the highest probability.

It is called **â€œNaÃ¯veâ€** because it **assumes that all features are independent of each other**, meaning one feature has no effect on another. This assumption is rarely true in real-world data (thus â€œnaÃ¯veâ€), but the algorithm still works surprisingly well and is very fast and effective, especially for text-based problems.


13. What is Bayesâ€™ Theorem

-->**Bayesâ€™ Theorem** is a fundamental rule in probability that describes how to update the probability of an event based on new evidence. It connects **prior probability**, **likelihood**, and **posterior probability**.

P(Aâˆ£B)=P(B|A)P(A)/P(B)â€‹


### **Meaning:**

* **(P(A|B))** â†’ Probability of event A *given* that B happened (**posterior**)
* **(P(B|A))** â†’ Probability of observing B *if* A is true (**likelihood**)
* **(P(A))** â†’ Probability of A before seeing evidence (**prior**)
* **(P(B))** â†’ Total probability of seeing evidence B (**normalization factor**)

16. What are the key assumptions made by NaÃ¯ve Bayes

--->The key assumptions made by **NaÃ¯ve Bayes** are:

---

## ðŸ”¹ **1. Feature Independence Assumption (NaÃ¯ve Assumption)**

NaÃ¯ve Bayes assumes that **all features are independent of each other** given the class label.
In other words,
[
P(X_1, X_2, ..., X_n | Y) = \prod_{i=1}^{n} P(X_i | Y)
]
This is rarely true in real-world data but simplifies computation and still works well.

---

## ðŸ”¹ **2. Equal Importance of All Features**

Each feature contributes **independently and equally** to determining the class.
There is no feature interaction considered.

---

## ðŸ”¹ **3. Correct Distributional Assumption**

Each variant assumes a certain probability distribution:

* **Gaussian NB:** features follow **normal distribution**
* **Multinomial NB:** features are **counts/frequencies**
* **Bernoulli NB:** features are **binary (0/1)**

---

## ðŸ”¹ **4. Predictors are Conditioned Only on the Class**

NaÃ¯ve Bayes assumes features depend **only on the class**, not on other features.

20. How does Laplace Smoothing help in NaÃ¯ve Bayes

-->**Laplace Smoothing** (also called **Add-One Smoothing**) helps NaÃ¯ve Bayes avoid the **zero-probability problem**, which occurs when a word or feature never appears in the training data for a particular class.

---

# ðŸ”¹ **Why the Problem Happens**

NaÃ¯ve Bayes multiplies probabilities of all features.
If one feature has probability **0**, the entire product becomes **0**, causing the model to misclassify.

Example:
If word â€œexcellentâ€ never appears in the *spam* class â†’
[
P(\text{excellent} | \text{spam}) = 0
]
Then the document can never be predicted as spam, even if other words strongly indicate spam.

---

#  **How Laplace Smoothing Fixes It**

Laplace Smoothing adds **1** to every count to ensure no probability becomes zero.

### Formula (Multinomial NaÃ¯ve Bayes):

[
P(w_i | c) = \frac{\text{count}(w_i,c) + 1}{\text{total words in }c + V}
]
where:

* (V) = vocabulary size
* (+1) ensures every word has at least a small probability

---

#  **Effect of Laplace Smoothing**

###Avoids zero probabilities

###Makes the model more robust

###Allows classification even when new/unseen words appear

###Stabilizes probability estimation

21.Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


iris = load_iris()
X = iris.data
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


svm_model = SVC(kernel='linear', random_state=42)


svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"SVM Classifier Accuracy on Iris Dataset: {accuracy:.4f}")
```

22. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then
compare their accuracies

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


wine = load_wine()
X = wine.data
y = wine.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


svm_linear = SVC(kernel='linear', random_state=42)
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print(f"SVM Classifier with Linear Kernel Accuracy: {accuracy_linear:.4f}")


svm_rbf = SVC(kernel='rbf', random_state=42)
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print(f"SVM Classifier with RBF Kernel Accuracy: {accuracy_rbf:.4f}")


if accuracy_linear > accuracy_rbf:
    print(f"\nLinear Kernel performed better by {accuracy_linear - accuracy_rbf:.4f}")
elif accuracy_rbf > accuracy_linear:
    print(f"\nRBF Kernel performed better by {accuracy_rbf - accuracy_linear:.4f}")
else:
    print("\nBoth Linear and RBF Kernels performed equally well.")
```

23.Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean
Squared Error (MSE)

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler


housing = fetch_california_housing()
X = housing.data
y = housing.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


svr_model = SVR(kernel='rbf')
svr_model.fit(X_train_scaled, y_train)


y_pred = svr_model.predict(X_test_scaled)


mse = mean_squared_error(y_test, y_pred)
print(f"SVR Mean Squared Error on Housing Dataset: {mse:.4f}")
```

24.Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


svm_poly = SVC(kernel='poly', degree=3, C=1, random_state=42)
svm_poly.fit(X_train_scaled, y_train)


x_min, x_max = X_train_scaled[:, 0].min() - 1, X_train_scaled[:, 0].max() + 1
y_min, y_max = X_train_scaled[:, 1].min() - 1, X_train_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))


Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)


plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train, cmap=plt.cm.viridis, edgecolors='k', s=80, label='Training points')

plt.scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], c=y_test, cmap=plt.cm.plasma, marker='x', s=100, label='Test points')

plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.title('SVM Classifier with Polynomial Kernel Decision Boundary (Iris Dataset)')
plt.legend()
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()
```

26.Write a Python program to train a Multinomial NaÃ¯ve Bayes classifier for text classification using the 20
Newsgroups dataset

```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report


categories = ['alt.atheism', 'soc.religion.christian',
              'comp.graphics', 'sci.med']
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=42)
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, shuffle=True, random_state=42)


X_train = newsgroups_train.data
y_train = newsgroups_train.target
X_test = newsgroups_test.data
y_test = newsgroups_test.target


vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)


mnb_model = MultinomialNB()
mnb_model.fit(X_train_counts, y_train)


y_pred = mnb_model.predict(X_test_counts)


accuracy = accuracy_score(y_test, y_pred)
print(f"Multinomial NaÃ¯ve Bayes Classifier Accuracy on 20 Newsgroups Dataset: {accuracy:.4f}")


print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=newsgroups_test.target_names))
```

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset (a good multi-class classification problem)
bcs = load_breast_cancer()
X = bcs.data
y = bcs.target

# Initialize StandardScaler for feature scaling
scaler = StandardScaler()

# Initialize SVM Classifier
# Using RBF kernel with default parameters, but these could be tuned with GridSearchCV within each fold
svm_model = SVC(kernel='rbf', random_state=42)

# Initialize Stratified K-Fold Cross-Validation
# n_splits determines the number of folds (e.g., 5 or 10 are common)
# shuffle=True ensures the data is shuffled before splitting
# random_state for reproducibility
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# List to store accuracy scores for each fold
accuracy_scores = []

print(f"Performing Stratified K-Fold Cross-Validation with {skf.get_n_splits(X, y)} splits...")

# Iterate over each fold
for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    print(f"\n--- Fold {fold + 1} ---")

    # Split data into training and test sets for the current fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Scale features. Fit scaler on training data, then transform both training and test data.
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train the SVM model on the scaled training data
    svm_model.fit(X_train_scaled, y_train)

    # Make predictions on the scaled test data
    y_pred = svm_model.predict(X_test_scaled)

    # Calculate and store the accuracy for the current fold
    fold_accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(fold_accuracy)
    print(f"Accuracy for Fold {fold + 1}: {fold_accuracy:.4f}")

# Compute and print the average accuracy across all folds
average_accuracy = np.mean(accuracy_scores)
print(f"\n----------------------------------------")
print(f"Average Accuracy across all folds: {average_accuracy:.4f}")
print(f"Standard Deviation of Accuracy across all folds: {np.std(accuracy_scores):.4f}")
```

40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE

# Load the Breast Cancer dataset
bcs = load_breast_cancer()
X = bcs.data
y = bcs.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features (important for SVMs)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- 1. Train SVM WITHOUT RFE (Baseline) ---
print("--- Training SVM WITHOUT RFE (Baseline) ---")
svm_no_rfe = SVC(kernel='linear', random_state=42) # Use linear kernel for RFE for coef_ attribute
svm_no_rfe.fit(X_train_scaled, y_train)
y_pred_no_rfe = svm_no_rfe.predict(X_test_scaled)
accuracy_no_rfe = accuracy_score(y_test, y_pred_no_rfe)
print(f"SVM Accuracy (Without RFE): {accuracy_no_rfe:.4f}")

# --- 2. Perform Recursive Feature Elimination (RFE) ---
print("\n--- Performing Recursive Feature Elimination (RFE) ---")
# RFE requires an estimator that exposes feature importances (e.g., coef_ for linear models)
# We'll use the same linear SVC as the estimator for RFE.
rfe_estimator = SVC(kernel='linear', random_state=42)

# Number of features to select (you can tune this, e.g., using GridSearchCV)
k_features_rfe = 10 # Let's select 10 features out of 30
rfe_selector = RFE(estimator=rfe_estimator, n_features_to_select=k_features_rfe, step=1)

# Fit RFE on the scaled training data
rfe_selector.fit(X_train_scaled, y_train)

# Transform the data to include only the selected features
X_train_rfe = rfe_selector.transform(X_train_scaled)
X_test_rfe = rfe_selector.transform(X_test_scaled)

print(f"Original number of features: {X.shape[1]}")
print(f"Number of features after RFE: {X_train_rfe.shape[1]}")

# Get the names of the selected features
selected_feature_indices_rfe = rfe_selector.get_support(indices=True)
selected_feature_names_rfe = [bcs.feature_names[i] for i in selected_feature_indices_rfe]
print("Selected features by RFE:", selected_feature_names_rfe)

# --- 3. Train SVM WITH RFE-selected Features ---
print("\n--- Training SVM WITH RFE-selected Features ---")
# Use an RBF kernel for the final SVM for potentially better performance
svm_with_rfe = SVC(kernel='rbf', random_state=42)
svm_with_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = svm_with_rfe.predict(X_test_rfe)
accuracy_rfe = accuracy_score(y_test, y_pred_rfe)
print(f"SVM Accuracy (With RFE): {accuracy_rfe:.4f}")

# --- 4. Compare Results ---
print("\n--- Comparison ---")
print(f"Accuracy WITHOUT RFE: {accuracy_no_rfe:.4f}")
print(f"Accuracy WITH RFE:    {accuracy_rfe:.4f}")

if accuracy_rfe > accuracy_no_rfe:
    print(f"RFE improved accuracy by: {accuracy_rfe - accuracy_no_rfe:.4f}")
elif accuracy_no_rfe > accuracy_rfe:
    print(f"Without RFE, accuracy was better by: {accuracy_no_rfe - accuracy_rfe:.4f}")
else:
    print("RFE did not change the accuracy significantly.")
```

46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay, average_precision_score
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset (a binary classification problem)
bcs = load_breast_cancer()
X = bcs.data
y = bcs.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features (important for SVMs)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train the SVM Classifier
# SVC does not directly output probabilities by default. It outputs a 'decision function'
# which can be used to compute precision/recall. For probability estimates, set probability=True
# but it adds computational cost (cross-validation on training data).
# For plotting PR curve, decision_function is often sufficient.
svm_model = SVC(kernel='rbf', random_state=42, probability=True) # Set probability=True for predict_proba
svm_model.fit(X_train_scaled, y_train)

# Get decision function scores or predicted probabilities for the positive class
# If probability=True, use predict_proba[:, 1]
# Otherwise, use decision_function()
y_scores = svm_model.predict_proba(X_test_scaled)[:, 1]

# Calculate precision, recall, and thresholds
precision, recall, _ = precision_recall_curve(y_test, y_scores)

# Calculate Average Precision (AP) score
avg_precision = average_precision_score(y_test, y_scores)

print(f"--- SVM Classifier Performance Evaluation ---")
print(f"Average Precision Score: {avg_precision:.4f}")

# Visualize the Precision-Recall curve
plt.figure(figsize=(8, 6))
display = PrecisionRecallDisplay(precision=precision, recall=recall, average_precision=avg_precision)
display.plot()
plt.title('Precision-Recall Curve for SVM Classifier (Breast Cancer Dataset)')
plt.grid(True)
plt.show()
```
